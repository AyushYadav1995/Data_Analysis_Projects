## Relationship between the Engine size, Cylinders and Hoursepower

```python
cars_df.head(2)
```

```python
sns.lineplot(x='EngineSize',y='Horsepower',hue='DriveTrain',data=cars_df)

plt.title('Size vs HoursePower');
```

### So for all type of DriveTrain there is lines relationship enginesize and horsepower

```python
sns.histplot(x='Horsepower',data=cars_df,bins=15);
```

### By this distribution most of the cars lies around 200 Horsepower

```python
sns.lineplot(x='Length',y='Weight',data=cars_df);

plt.title('Lenth vs weight')
```

### Now it would be better if visualize by grouping 

```python
cars_make_df=cars_df.groupby('Make').mean()
```

```python
cars_make_df.reset_index(inplace=True)
```

```python
cars_make_df.head()
```

```python
sns.scatterplot(x='Make',y='EngineSize',data=cars_make_df)

plt.xlabel('Cars models')

plt.xticks(rotation=90)


plt.title('CARS SIZE');
```

## So Here we found following abswers:
### 1- Highest engine size car= Hummer
### 2- Number of cars have their engine size above 4 = Cadillac,GMC,Hummer and Lincoln

```python
fig,axes=plt.subplots(2,2)

#jan line plot day wise
jan_day_df=wh_df[wh_df['month']==1].groupby('day').mean()
jan_day_df.reset_index(inplace=True)

axes[0,0].plot(jan_day_df['day'],jan_day_df['Temp_C'])

axes[0,0].set_title('jan temp');

axes[0,0].set_xlabel('Days')
axes[0,0].set_ylabel('Temp in C')


# jan scatterplot day wise
sns.scatterplot(x='day',y='Temp_C',data=jan_day_df,ax=axes[1,0])

axes[1,0].set_title('jan day scatter temp')

axes[1,0].set_xlabel('Days')
axes[1,0].set_ylabel('Temp in C')


# jan Bar plot weekday wise
jan_weekday_df=wh_df[wh_df['month']==1].groupby('weekday').mean()
jan_weekday_df.reset_index(inplace=True)

sns.barplot(x='weekday',y='Temp_C',data=jan_weekday_df, ax=axes[0,1])

axes[0,1].set_title('jan weekday temp');

axes[0,1].set_xlabel('Days')
axes[0,1].set_ylabel('Temp in C')


# Histogram(distribution chart)
sns.histplot(x='Temp_C',data=jan_day_df,ax=axes[1,1])

axes[1,1].set_title('Distribtion of temp')

axes[1,1].set_xlabel('Temp in C')
axes[1,1].set_ylabel('percentage')



plt.tight_layout(pad=0.5)
```
